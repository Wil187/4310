```python
import pandas as pd
```

## Exercise 1

```python
faang = pd.DataFrame()
for ticker in ['fb', 'aapl', 'amzn', 'nflx', 'goog']:
    df = pd.read_csv(f'../ch_03/exercises/{ticker}.csv')
    df.insert(0, 'ticker', ticker.upper())
    faang = faang.append(df)

faang.to_csv('faang.csv', index=False)
```

## Exercise 2

```python
faang['date'] = pd.to_datetime(faang['date'])
faang['volume'] = faang['volume'].astype(int)
faang.sort_values(by=['ticker','date'])


faang[5:]
```

## Exercise 3

```python
faang.nsmallest(7, 'volume')
```

## Exercise 6

```python
covid = pd.read_csv('../ch_03/exercises/covid19_cases.csv').assign(
    date=lambda x: pd.to_datetime(x.dateRep, format='%d/%m/%Y')).set_index('date')
covid.replace('United_States_of_America', 'USA')
covid.replace('United_Kingdom', 'UK')
covid.sort_index()

covid[covid.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])
     ].reset_index().pivot(index='date', columns='countriesAndTerritories', values='cases').fillna(0)
```
